### Finding outliers


Find out if outliers exist in the column `ConvertedComp` using a box plot?

Find out the Inter Quartile Range for the column `ConvertedComp`.


```python
# your code goes here
df.ConvertedComp.describe()
```

```python
df.ConvertedComp.median()
```

```python
1.000000e+05 - 2.686800e+04
```

Find out the upper and lower bounds.


```python
# your code goes here
Quartile_1=df.ConvertedComp.quantile(.25)
Quartile_3=df.ConvertedComp.quantile(.75)
IQR=Quartile_3-Quartile_1
IQR
```

Identify how many outliers are there in the `ConvertedComp` column.


```python
# your code goes here
outliers = (df.ConvertedComp < (Quartile_1 - 1.5 * IQR)) | (df.ConvertedComp > (Quartile_3 + 1.5 * IQR))
outliers.value_counts()
```

Create a new dataframe by removing the outliers from the `ConvertedComp` column.


```python
# your code goes here
lower=(df.ConvertedComp < (Quartile_1 - 1.5 * IQR))
upper=(df.ConvertedComp > (Quartile_3 + 1.5 * IQR))
```

```python
lower.value_counts()
```

```python
upper.value_counts()
```

```python
ConvertedComp_Out_free=df[~(df.ConvertedComp > (Quartile_3 +1.5*IQR))]
```

```python
ConvertedComp_Out_free.head(4)
```

```python
ConvertedComp_Out_free.ConvertedComp.median()
```

```python
ConvertedComp_Out_free.ConvertedComp.mean()
```
